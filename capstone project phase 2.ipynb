import pandas as pd

import seaborn as sns

 Load and Explore the Data

```python
# Load your dataset
data = {
    'State/UT': ['State1', 'State2', 'State3'],
    'Population': [12345678, 87654321, 23456789],
    'Literacy Rate': [88.7, 76.5, 82.3],
    'Total Crimes': [4500, 7800, 3300],
    'Crime Type1': [1500, 2500, 1000],
    'Crime Type2': [1200, 3200, 900],
    'Crime Type3': [1800, 2100, 1400],
    'Year': [2021, 2021, 2021],
    'Area': [45000, 67000, 54000]
}

df = pd.DataFrame(data)
print(df.head())
print(df.info())
print(df.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Literacy Rate', y='Total Crimes', data=df)
plt.title('Literacy Rate vs Total Crimes')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()
```

Analysis of the Type of Crime vs Each State vs Literacy Rate
python


```python
crime_types = ['Crime Type1', 'Crime Type2', 'Crime Type3']
df_melt = df.melt(id_vars=['State/UT', 'Literacy Rate'], value_vars=crime_types, var_name='Crime Type', value_name='Crime Count')

plt.figure(figsize=(12, 8))
sns.barplot(x='State/UT', y='Crime Count', hue='Crime Type', data=df_melt)
plt.title('Type of Crime vs Each State vs Literacy Rate')
plt.xlabel('State/UT')
plt.ylabel('Crime Count')
plt.grid(True)
plt.show()
```

Observations:

This bar plot visualizes different types of crimes across each state, along with the literacy rate.
Differences in crime rates across states and crime types can be analyzed.


 Analysis of Year-on-Year Total Crime Rate

```python
# Assuming we have data for multiple years
yearly_data = {
    'State/UT': ['State1', 'State1', 'State2', 'State2', 'State3', 'State3'],
    'Year': [2020, 2021, 2020, 2021, 2020, 2021],
    'Total Crimes': [4000, 4500, 7500, 7800, 3100, 3300]
}
df_yearly = pd.DataFrame(yearly_data)

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Total Crimes', hue='State/UT', data=df_yearly, marker='o')
plt.title('Year-on-Year Total Crime Rate')
plt.xlabel('Year')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()
```

 Analysis of Area vs Overall Crime

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Population', y='Total Crimes', data=df)
plt.title('Population vs Overall Crime')
plt.xlabel('Population')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()
```
